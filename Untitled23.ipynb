```python
from sklearn.model_selection import train_test_split
train_set , test_set = train_test_split(df,random_state=21,train_size=0.8)
```

```python
uylar = train_set.drop(['location','district','price'],axis=1)
Y = train_set["price"]
```

```python
# ma'lumotlarni normalizatsiyaga keltirib olishimiz kerak
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
data_scaler = std_scaler.fit_transform(uylar)
```

**Modellaymiz**

Linear Regression 

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
```

```python
LR_model.fit(data_scaler,Y)
```

```python
#modelimiz tayyor!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!\
```

**Modelni baholaymiz**

```python
test_data = uylar.sample(5)
```

```python
real_qiymat=Y.loc[test_data.index]
real_qiymat  # uylarning asl bahosi
```

```python
test_data_scaler = std_scaler.transform(test_data)
```

```python
predict_test_data = LR_model.predict(test_data_scaler)
predict_test_data
```

```python
pd.DataFrame({"real":real_qiymat,"bashorat":predict_test_data})
```

```python
#endi boshida ikkiga bo'lgan test datani ishlatib ko'ramiz
housing2 = test_set.drop(['location',"district",'price'],axis=1)
test_label = test_set["price"]
```

```python
test_label.shape
```

```python
test_set_scaler = std_scaler.fit_transform(housing2)
```

```python
predict_test_set = LR_model.predict(test_set_scaler)
```

```python
predict_test_set.shape
```

```python
from sklearn.metrics import mean_squared_error
qiymat = mean_squared_error(predict_test_set,test_label)
javob = np.sqrt(qiymat)
print(f"RMAE={javob}")
```
